```python
import pandas as pd
import numpy as np
```

```python
hw = pd.read_csv("../Dataset/height-weight.csv")
hw
```

```python
weight_mean = hw["Weight"].mean()
weight_mean
```

```python
height_mean = hw["Height"].mean()
height_mean
```

```python
hw["weight-mean(w)"] = hw["Weight"] - weight_mean
hw
```

```python
hw["height-mean(h)"] = hw["Height"] - height_mean
hw
```

```python
hw["weight-mean(w)-square"] = hw["weight-mean(w)"] * hw["weight-mean(w)"]
hw
```

```python
hw['Height-weight-mul'] = hw['height-mean(h)'] * hw['weight-mean(w)']
hw
```

```python
sum_mul = hw['Height-weight-mul'].sum()
sum_mul
```

```python
sum_square = hw['weight-mean(w)-square'].sum()
sum_square
```

```python
b1 = sum_mul/sum_square
b1
```

```python
b0 = height_mean + (b1*weight_mean)
b0
```

```python
hw['y cap']= b0 + (b1*hw['Weight'])
hw
```
